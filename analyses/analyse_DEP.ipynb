```python
# Imports
import pandas as pd
```

```python
# Datasets inladen
klassement = pd.read_csv("D:/Hogent/Visual Studio Code/DEP/DEP1-2023-2024-groep30/transfermarkt/data/correcte_data/klassement.csv", sep=";")
```

```python
klassement.head(20)
```

```python
def segment_bepalen(positie, aantal_teams):
    if positie <= (aantal_teams / 2):
        return "Boven"  
    else:
        return "Onder"

def verwerken(klassement):
    # Aantal teams in de competitie berekenen
    aantal_teams_per_seizoen = klassement.groupby('Seizoen')['Stamnummer'].nunique().to_dict()

    # berekenen van deel klassement
    klassement['PositiePloeg'] = klassement.apply(
        lambda x: segment_bepalen(x['Stand'], aantal_teams_per_seizoen[x['Seizoen']]), axis=1
    )

    return klassement

wedstrijden = verwerken(klassement)
```

```python
klassement.head(20)
```

```python
# lege lijst
aantal_keer_bovenaan = []

# Loop door elk uniek seizoen in de dataset
for seizoen in klassement['Seizoen'].unique():
    # Selecteer huidig seizoen
    seizoens_gegevens = klassement[klassement['Seizoen'] == seizoen]
    
    # Selecteer laaste speeldag
    laatste_speeldag = seizoens_gegevens['Speeldag'].max()
    
    # Selecteer laatste speeldag van het huidige seizoen
    laatste_speeldag_gegevens = seizoens_gegevens[seizoens_gegevens['Speeldag'] == laatste_speeldag]
    
    # Tel hoe vaak elke ploeg Boven staat 
    aantal_keer_bovenaan_per_seizoen = laatste_speeldag_gegevens['Ploeg'][laatste_speeldag_gegevens['PositiePloeg'] == 'Boven'].value_counts()
    
    # Voeg de resultaten toe aan de lijst
    aantal_keer_bovenaan.append(aantal_keer_bovenaan_per_seizoen)

# Combineer de resultaten van elk seizoen tot één DataFrame
aantal_keer_bovenaan = pd.concat(aantal_keer_bovenaan, axis=1).sum(axis=1).reset_index()
aantal_keer_bovenaan.columns = ['Ploeg', 'Aantal_keer_bovenaan']

# Bekijk de resultaten
print(aantal_keer_bovenaan)
```

```python
sorted_aantal_keer_bovenaan = aantal_keer_bovenaan.sort_values(by='Aantal_keer_bovenaan', ascending=False)
```
